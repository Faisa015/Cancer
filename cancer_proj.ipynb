```python
df.drop(['id','Unnamed: 32'], axis=1, inplace=True)
df
```

```python
x = df.drop(['diagnosis'],axis=1)
y = df['diagnosis']
x_train,x_test,y_train,y_test = train_test_split(x,y ,test_size=0.20,random_state=24)
```

```python
R=KNeighborsClassifier(n_neighbors=1)
R.fit(x_train,y_train)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
pred
```

```python
rv = np.array(y_test)
rv
```

```python
accuracy_sklearn = accuracy_score(y_test,pred)*100
print(f'model accuracy: {accuracy_sklearn}')
```

```python
svm = SVC()
svm.fit(x_train,y_train)
```

```python
print("Accuracy of SVM classifier on traninig set: {}"
      .format(svm.score(x_train, y_train)))
print("Accuracy of SVM classifier on traninig set: {}"
      .format(svm.score(x_test, y_test)))
      
```

```python
predicted_labels = svm.predict(x_test)
```
